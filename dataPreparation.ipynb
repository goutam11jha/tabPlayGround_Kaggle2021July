```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df['date_time'] = pd.to_datetime(df['date_time'])
df.columns
df[['sensor_1', 'sensor_2','sensor_3','sensor_4','sensor_5']].corr()
```

```python
df.set_index('date_time', inplace=True)
df['per']=df.index.to_period('M')
```

```python
df[['target_carbon_monoxide','per']].boxplot(by='per')
plt.xticks(rotation=90)
plt.title('CO variation on month')
plt.savefig("CO_var.jpg")
```

```python
df[['target_benzene','per']].boxplot(by='per')
plt.xticks(rotation=90)
plt.title('Benzene variation on month')
plt.savefig("Benzene_var.jpg")
```

```python
df[['target_nitrogen_oxides','per']].boxplot(by='per')
plt.xticks(rotation=90)
plt.title('NO variation on month')
plt.savefig("NO_var.jpg")
```
